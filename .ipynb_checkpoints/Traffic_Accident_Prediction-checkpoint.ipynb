```python
# Cell 2
def normalize_state_name(name):
    if pd.isna(name): return name
    name = str(name).lower().strip()
    return re.sub(r'[^a-z\s]', '', name)

print("Loading and Merging datasets...")
df_india = pd.read_csv("data/accident_prediction_india.csv")
df_driver = pd.read_csv("data/driverresponse.csv")
df_ncrb = pd.read_csv("data/NCRB_ADSI_2023_Table_1A.7.csv")
df_road = pd.read_csv("data/Road.csv")

# Normalize States
df_india['Normalized_State'] = df_india['State Name'].apply(normalize_state_name)
df_ncrb['Normalized_State'] = df_ncrb['State/UT'].apply(normalize_state_name)
df_driver['Normalized_State'] = df_driver['stateut'].apply(normalize_state_name)

# Merge 1, 2, 3
merged_df = pd.merge(df_india, df_ncrb[['Normalized_State', 'Total - Cases', 'Total - Injured', 'Total - Died']], on='Normalized_State', how='left')
driver_cols = ['Normalized_State', 'alcintake2016', 'overspeed2016', 'wrongside2016', 'lanejumping2016']
avail_driver_cols = [c for c in driver_cols if c in df_driver.columns]
if 'Normalized_State' not in avail_driver_cols: avail_driver_cols.insert(0, 'Normalized_State')
merged_df = pd.merge(merged_df, df_driver[avail_driver_cols], on='Normalized_State', how='left')

# Align and stack Road.csv
road_mapping = {
    'Time': 'Time of Day', 'Day_of_week': 'Day of Week', 'Accident_severity': 'Accident Severity',
    'Weather_conditions': 'Weather Conditions', 'Number_of_vehicles_involved': 'Number of Vehicles Involved',
    'Type_of_vehicle': 'Vehicle Type Involved', 'Number_of_casualties': 'Number of Casualties',
    'Road_surface_conditions': 'Road Condition', 'Light_conditions': 'Lighting Conditions',
    'Sex_of_driver': 'Driver Gender', 'Age_band_of_driver': 'Driver Age'
}
df_road_aligned = df_road.rename(columns=road_mapping)
common_cols = list(set(merged_df.columns).intersection(set(df_road_aligned.columns)))
final_df = pd.concat([merged_df, df_road_aligned[common_cols]], axis=0, ignore_index=True)

if 'Normalized_State' in final_df.columns: final_df.drop(columns=['Normalized_State'], inplace=True)
final_df.to_csv("outputs/merged_accident_data.csv", index=False)
print(f"Data successfully merged! Shape: {final_df.shape}")
final_df.head() # Display the first few rows in Jupyter
```

```python
import pandas as pd
import re
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE

print("Preprocessing Data...")
df = pd.read_csv("outputs/merged_accident_data.csv").dropna(subset=['Accident Severity'])

def extract_hour(time_str):
    try: return int(str(time_str).split(':')[0])
    except: return np.nan

df['Hour'] = df['Time of Day'].apply(extract_hour).fillna(12)
df['Is_Night'] = df['Hour'].apply(lambda x: 1 if (x >= 18 or x <= 6) else 0)
df['Is_Weekend'] = df['Day of Week'].apply(lambda d: 1 if str(d).lower() in ['saturday', 'sunday'] else 0)

severity_map = {'Slight Injury': 0, 'Minor': 0, 'Serious Injury': 1, 'Serious': 1, 'Fatal injury': 2, 'Fatal': 2}
df['Target_Severity'] = df['Accident Severity'].map(severity_map)
df.dropna(subset=['Target_Severity'], inplace=True)
y = df['Target_Severity'].astype(int)

# --- THE SMART FILTER ---
desired_features = [
    'State Name', 'City Name', 'Day of Week', 'Hour', 'Is_Night', 'Is_Weekend', 
    'Weather Conditions', 'Road Condition', 'Vehicle Type Involved', 
    'Number of Vehicles Involved', 'Speed Limit (km/h)',
    'Road_surface_type', 'Area_accident_occured', 'Type_of_collision', 
    'Light_conditions', 'Road_allignment', 'Road_Type' 
]

# Only keep the ones that ACTUALLY exist in your dataset to prevent crashes!
features_to_keep = [col for col in desired_features if col in df.columns]
missing_features = [col for col in desired_features if col not in df.columns]

print(f"✅ Successfully found {len(features_to_keep)} columns to train on.")
if missing_features:
    print(f"⚠️ Warning: Skipped these columns because they are not in your dataset: {missing_features}")

X = df[features_to_keep].copy()

# Handle Missing values
for col in X.select_dtypes(include=['float64', 'int64']).columns: 
    X[col].fillna(X[col].median(), inplace=True)
for col in X.select_dtypes(include=['object']).columns: 
    X[col].fillna("Unknown", inplace=True) 

# Encode
label_encoders = {}
for col in X.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le
    
joblib.dump(label_encoders, 'outputs/models/label_encoders.pkl')
joblib.dump(X.columns.tolist(), 'outputs/models/feature_columns.pkl')

# Scale & Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
joblib.dump(scaler, 'outputs/models/scaler.pkl')

# SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)

joblib.dump((X_train_res, X_test_scaled, y_train_res, y_test), 'outputs/models/processed_data.pkl')
print("Success! Preprocessing complete.")
```

Step 3 - Train Models

```python
import joblib
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import f1_score

print("Training Base Models...")
X_train, X_test, y_train, y_test = joblib.load('outputs/models/processed_data.pkl')

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42, verbose=-1)
}

best_f1, best_model, best_model_name = 0, None, ""

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    f1 = f1_score(y_test, y_pred, average='weighted')
    print(f"{name} F1-Score: {f1:.4f}")
    if f1 > best_f1:
        best_f1, best_model, best_model_name = f1, model, name

print(f"\nBest Model: {best_model_name}")
joblib.dump(best_model, 'outputs/models/best_model.pkl')
```

Step 4 - Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix

# Instead of just saving them, we use plt.show() to render them directly in the notebook!
best_model = joblib.load('outputs/models/best_model.pkl')
features = joblib.load('outputs/models/feature_columns.pkl')

# 1. Feature Importance (If it's a tree model like XGBoost/RF)
if hasattr(best_model, 'feature_importances_'):
    importances = best_model.feature_importances_
    indices = np.argsort(importances)[-15:]
    
    plt.figure(figsize=(10, 6))
    plt.title('Top 15 Feature Importances')
    plt.barh(range(len(indices)), importances[indices], color='b', align='center')
    plt.yticks(range(len(indices)), [features[i] for i in indices])
    plt.xlabel('Relative Importance')
    plt.show()

# 2. Confusion Matrix
y_pred = best_model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

Step 5 - Prediction Test

```python
def predict_severity_jupyter(sample_dict):
    best_model = joblib.load('outputs/models/best_model.pkl')
    scaler = joblib.load('outputs/models/scaler.pkl')
    label_encoders = joblib.load('outputs/models/label_encoders.pkl')
    feature_columns = joblib.load('outputs/models/feature_columns.pkl')

    df_sample = pd.DataFrame([sample_dict])
    for col in feature_columns:
        if col not in df_sample.columns: df_sample[col] = 0

    df_sample = df_sample[feature_columns]

    for col, le in label_encoders.items():
        if col in df_sample.columns:
            df_sample[col] = df_sample[col].apply(lambda x: le.transform([str(x)])[0] if str(x) in le.classes_ else 0)

    X_scaled = scaler.transform(df_sample)
    pred_class = best_model.predict(X_scaled)[0]
    probs = best_model.predict_proba(X_scaled)[0]
    
    mapping = {0: 'Minor/Slight', 1: 'Serious', 2: 'Fatal'}
    print(f"Predicted Severity: {mapping.get(pred_class, 'Unknown')} (Class {pred_class})")
    print(f"Probabilities: Minor: {probs[0]:.2f}, Serious: {probs[1]:.2f}, Fatal: {probs[2]:.2f}")

# Test the prediction
sample_input = {
    'Day of Week': 'Saturday',
    'Hour': 23,
    'Is_Night': 1,
    'Weather Conditions': 'Rainy',
    'Road Condition': 'Wet',
    'Vehicle Type Involved': 'Truck',
    'Number of Vehicles Involved': 2,
    'Speed Limit (km/h)': 80
}

predict_severity_jupyter(sample_input)
```
